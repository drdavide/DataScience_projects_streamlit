```python
import pandas as pd
def load_data():
    url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    html = pd.read_html(url, header=0)
    df = html[0]
    return df

df = load_data()
df
```

```python
# source code from https://pypi.org/project/yfinance/

data = yf.download(  # or pdr.get_data_yahoo(...
        # tickers list or string as well
        # e.g.,tickers = "SPY AAPL MSFT",
        tickers = list(df.Symbol),

        # use "period" instead of start/end
        # valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
        # (optional, default is '1mo')
        # e.g., period = "ytd",
        period = "ytd",

        # fetch data by interval (including intraday if period < 60 days)
        # valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
        # (optional, default is '1d')
        interval = "1d",

        # group by ticker (to access via data['SPY'])
        # (optional, default is 'column')
        group_by = 'ticker',

        # adjust all OHLC automatically
        # (optional, default is False)
        auto_adjust = True,

        # download pre/post regular market hours data
        # (optional, default is False)
        prepost = True,

        # use threads for mass downloading? (True/False/Integer)
        # (optional, default is True)
        threads = True,

        # proxy URL scheme use use when downloading?
        # (optional, default is None)
        proxy = None
    )
```

```python
# we want make the date one of the column, together with the closing price (you can choose what you want to see)
newdf = pd.DataFrame(data['AAPL'].Close)
newdf['Date'] = newdf.index
newdf
```

```python
# Plot the closing price
import matplotlib.pyplot as plt
plt.fill_between(newdf.Date, newdf.Close, color="skyblue", alpha=0.3)
plt.plot(newdf.Date, newdf.Close, color="skyblue", alpha=0.8)
plt.xticks(rotation=90)
plt.xlabel("Date", fontweight="bold")
plt.ylabel("Closing Price", fontweight="bold")
```

```python
# We can create a function to plot the price of the stock of interest
def price_plot(symbol):
    df = pd.DataFrame(data[symbol].Close)
    df["Date"] = df.index
    plt.fill_between(df.Date, df.Close, color="skyblue", alpha=0.3)
    plt.plot(df.Date, df.Close, color="skyblue", alpha=0.8)
    plt.xticks(rotation=90)
    plt.title(symbol, fontweight="bold")
    plt.xlabel("Date", fontweight="bold")
    plt.ylabel("Closing Price", fontweight="bold")
    return plt.show()
```

```python
price_plot('ABT')
```

```python
# plot the price of the first 10 companies
for i in list(df.Symbol)[:10]:
    price_plot(i)
```
